```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import collections

#Creating a list of data
data = [i for i in range(-500,500)]
print(data)
    
```

```python
#counting the repeated values in the data set
count = collections.Counter(data)
print(count)
```

```python
#calculating frequency
for k,v in count.iteritems():
    print("The frequency of "+str(k)+" is "+str(float(v)/len(data)))
```

```python
plt.title('Box Plot')
plt.boxplot(data)
plt.show()
```

```python
#Creating a histogram
plt.title('Histogram')
plt.hist(data, histtype='bar')
```

```python
#Creating a Q-Q plot
plt.figure()
graph = stats.probplot(data, dist="norm", plot=plt)
```
